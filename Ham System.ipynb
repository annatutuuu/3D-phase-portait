```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def system(t, z):
    x, y = z
    dxdt = y
    dydt = x - x**3
    return [dxdt, dydt]

def plot_phase_portrait(system, x_range, y_range, num_points=20):
    x = np.linspace(x_range[0], x_range[1], num_points)
    y = np.linspace(y_range[0], y_range[1], num_points)

    X, Y = np.meshgrid(x, y)
    U, V = np.zeros(X.shape), np.zeros(Y.shape)

    for i in range(num_points):
        for j in range(num_points):
            xdot, ydot = system(0, [X[i, j], Y[i, j]])
            U[i, j] = xdot
            V[i, j] = ydot

    plt.streamplot(X, Y, U, V, color='blue')
    plt.xlim(x_range)
    plt.ylim(y_range)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Phase Portrait')
    plt.grid(True)
    plt.show()

# Define the range for x and y values
x_range = [-2, 2]
y_range = [-2, 2]

# Call the function to plot the phase portrait
plot_phase_portrait(system, x_range, y_range)
```
